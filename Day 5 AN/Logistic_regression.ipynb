```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv(r'C:\Users\tharu\OneDrive\Desktop\Data_Scientist\Day 5 AN\framingham_heart_disease.csv')
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df = df.drop('education',axis=1)
df
```

```python
cig_mean = round(df['cigsPerDay'].mean())
```

```python
df['cigsPerDay'].fillna(cig_mean,inplace=True)
df.isnull().sum()
```

```python
bp_mean = round(df['BPMeds'].mean())
bp_mean
```

```python
df['BPMeds'].fillna(bp_mean,inplace=True)
df.isnull().sum()
```

```python
totchol_mean = round(df['BPMeds'].mean())
totchol_mean
```

```python
df['totChol'].fillna(totchol_mean,inplace=True)
df.isnull().sum()
```

```python
BMI_mean = round(df['BMI'].mean())
BMI_mean
```

```python
df['BMI'].fillna(BMI_mean,inplace=True)
df.isnull().sum()
```

```python
heartRate_mean = round(df['heartRate'].mean())
heartRate_mean
```

```python
df['heartRate'].fillna(heartRate_mean,inplace=True)
df.isnull().sum()
```

```python
glucose_mean = round(df['glucose'].mean())
glucose_mean
```

```python
df['glucose'].fillna(glucose_mean,inplace=True)
df.isnull().sum()
```

```python
x = df.drop('TenYearCHD',axis=1)
x
y = df['TenYearCHD']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=42)
```

```python
df.isnull().sum()
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_predicted = model.predict(x_train)
x_train_predicted
```

```python
train_accuracy = accuracy_score(y_train, x_train_predicted)
train_accuracy
```

```python
x_test_predicted = model.predict(x_test)
x_test_predicted
```

```python
test_accuracy = accuracy_score(y_test, x_test_predicted)
test_accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, x_test_predicted))
```

```python
cm = confusion_matrix(y_test, x_test_predicted)
cm
```
