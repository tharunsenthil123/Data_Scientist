```python
#machine learning for probability


import numpy as np

# 1. Define the given probabilities
p_disease = 0.01            # P(D): Probability of having the disease (Prior)
p_no_disease = 1 - p_disease # P(D^c): Probability of not having the disease

p_T_given_D = 0.95          # P(T|D): True Positive Rate
p_T_given_not_D = 0.05      # P(T|D^c): False Positive Rate

# 2. Calculate the total probability of a positive test, P(T)
# P(T) = P(T|D) * P(D) + P(T|D^c) * P(D^c)
p_T = (p_T_given_D * p_disease) + (p_T_given_not_D * p_no_disease)

# 3. Apply Bayes' Theorem to find P(D|T)
# P(D|T) = (P(T|D) * P(D)) / P(T)
p_D_given_T = (p_T_given_D * p_disease) / p_T

# Output the results
print(f"P(D) (Prior Probability): {p_disease}")
print(f"P(T|D) (True Positive Rate): {p_T_given_D}")
print(f"P(T|D^c) (False Positive Rate): {p_T_given_not_D}")
print(f"P(T) (Total Probability of Positive Test): {p_T:.6f}")
print(f"P(D|T) (Posterior Probability of Disease given Positive Test): {p_D_given_T:.4f}")
```
